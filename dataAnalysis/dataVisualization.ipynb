```python
past_present_count = pd.read_csv('past_present_count.csv', encoding='cp949')
past_present_count.head()
```

```python
seoul_regions = pd.unique(past_present_count['지역(구)'])
seoul_regions
```

```python
categories = ['경양식','기타','김밥(도시락)','까페','분식','뷔페식','식육(숯불구이)','일식','정종/대포집/소주방','중국식','통닭(치킨)','패밀리레스트랑','패스트푸드','한식','외국음식전문점(인도태국등)','횟집']
```

```python
plt.figure(figsize = (96, 48))
a = 1
for c in categories:
    plt.subplot(4, 4, a)
    plt.title(c)
    sns.barplot(x = '지역(구)', y = c, hue = '코로나이전여부', data = past_present_count)
    a += 1
```
